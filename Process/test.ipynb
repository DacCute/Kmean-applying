```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('../Data/e-shop clothing 2008.csv',sep=';')
```

```python
data
```

```python
df = pd.DataFrame(data.groupby('session ID')[['session ID', 'country', 'price']].mean())
df[['session ID', 'country']] = df[['session ID', 'country']].astype(int)
df.rename(columns={'session ID': 'custumers'}, inplace= True)
df[['trousers', 'skirts', 'blouses', 'sale']] = 0  
```

```python
df['price'] = data.groupby('session ID')['price'].mean()
```

```python
page_1 = ['', 'trousers', 'skirts', 'blouses', 'sale']
```

```python
data['page 1 (main category)'].iloc[165473]
```

```python
cr = 0
for i, row in df.iterrows():
    print(i,end= ': ')
    while data['session ID'].iloc[cr] == row['custumers']:
        print(cr, end = ' ')
        row[page_1[data['page 1 (main category)'].iloc[cr]]]  += 1
        cr += 1
        if cr == 165474:
            break
    df.loc[i,['trousers', 'skirts', 'blouses', 'sale']] = row[['trousers', 'skirts', 'blouses', 'sale']]
    print(" ")
    
```

```python
df
```

```python
choose_feature = data[["session ID", "page 1 (main category)", "country", "colour", "price"]]
```

```python
# Initialize the KMeans model
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
```

```python
kmeans.fit(df)
```

```python
clustered_dfs = []

for i in range(kmeans.n_clusters):
    cluster_members = df[kmeans.labels_ == i]
    clustered_dfs.append(cluster_members)

clustered_dfs
```

```python
centroids = kmeans.cluster_centers_
centroids
```
