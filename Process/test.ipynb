```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlot', 'Charlie'],
    'age': [25, 30, 29, 22]
})

# Define the new row (you can adjust the values)
new_row = {'name': 'David', 'age': 28}

# Create a DataFrame with the new row
new_df = pd.DataFrame([new_row])

# Replace rows where age is greater than 25 by concatenating new_df
condition = df['age'] > 25
df = pd.concat([df[~condition], new_df], ignore_index=True)

df
```

```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'one': [1, 1, 1],
    'two': [1, 2, 1]
})

# Group by all columns and count occurrences
counts = df.groupby(df.columns.tolist(), as_index=False).size()

print(counts)
type(counts)
```

```python
# Example DataFrames
df = pd.DataFrame({
    'team': ['A', 'B', 'C', 'D'],
    'score': [10, 20, 30, 40]
})

df2 = pd.DataFrame({
    'team': ['A', 'C'],
    'score': [100, 300]
})

# Replace rows in df where team is 'A' with corresponding rows from df2
df.loc[df['team'] == 'A'] = df2

print(df)
```
